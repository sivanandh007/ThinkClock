```python
import pandas as pd
import plotly.express as px

# Load the dataset
  # Update with the correct path to your file
data = pd.read_csv('metadata.csv')

# Inspect the first few rows of the data
print(data.head())

# Check the summary of the 'Capacity' column
print(data['Capacity'].describe())

# Check for missing values in 'Capacity'
print(f"Missing values in 'Capacity' column: {data['Capacity'].isna().sum()}")

# Convert 'Capacity' column to numeric (invalid values become NaN)
data['Capacity'] = pd.to_numeric(data['Capacity'], errors='coerce')

# Check the number of valid rows in the 'Capacity' column after conversion
print(f"Valid rows in 'Capacity' column: {data['Capacity'].notna().sum()}")

# Drop rows with missing 'Capacity' values
data_cleaned = data.dropna(subset=['Capacity'])

# Check if there is any data left to plot
print(f"Rows after cleaning: {data_cleaned.shape[0]}")

# Create a line plot for Capacity vs Aging (Test ID)
fig_capacity = px.line(data_cleaned, x='test_id', y='Capacity', title='Capacity vs Aging (Test ID)',
                       labels={'Capacity': 'Capacity (Ah)', 'test_id': 'Test ID'})

# Show the plot
fig_capacity.show()
```
